```python
@numba.jit
def min_in_range_numba(wvfm,t_start,t_min,t_max):
    time_arr = np.linspace(start=t_start,stop=t_start+0.002*np.size(wvfm),num=np.size(wvfm),endpoint=False)
    return np.amin(np.ma.masked_where((time_arr<t_min)^(time_arr>t_max),wvfm))

def min_in_range(row,t_min,t_max):
    return min_in_range_numba(row["wvfm"],row["time"],t_min,t_max)

@numba.jit
def min_in_range_time_numba(wvfm,t_start,t_min,t_max):
    time_arr = np.linspace(start=t_start,stop=t_start+0.002*np.size(wvfm),num=np.size(wvfm),endpoint=False)
    return np.argmin(np.ma.masked_where((time_arr<t_min)^(time_arr>t_max),wvfm))*0.002+t_start

def min_in_range_time(row,t_min,t_max):
    return min_in_range_time_numba(row["wvfm"],row["time"],t_min,t_max)
```
