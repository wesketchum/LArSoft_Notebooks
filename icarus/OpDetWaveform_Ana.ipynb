```python
from gallery_tools import *
```

```python
opdets_tag = art.InputTag("opdaq::DetSim");
filenames = ROOT.vector(ROOT.string)()
for line in open("/icarus/data/users/icaruspro/archive_work_Aug2018/corsika_detsim_list_Oct8.txt"):
    filenames.push_back(line.strip())
```

```python
filenames[0]
```

```python
ev = gallery.Event(filenames)
get_opdets = ev.getValidHandle(ROOT.vector(raw.OpDetWaveform))
```

```python
df_columns = ["run","event","opdet_idx","ch","time","wvfm"]
opdet_df = pd.DataFrame(columns=df_columns)

outfile_hdf = "/icarus/data/users/wketchum/corsika_detsim_opdet_df_trim.hdf"
```

```python
max_events=1000
event_counter=0

events_per_key=100
key_number=1
verbose=True

lowrange_us = -200.
hirange_us = 200.

while event_counter<max_events and (not ev.atEnd()):
    if verbose: print "Processing event %d" % event_counter

    this_opdet_df = pd.DataFrame(columns=df_columns)
    
    run = ev.eventAuxiliary().run()
    event = ev.eventAuxiliary().event()
        
    opdets = get_opdets(opdets_tag)
    if verbose: print "\tThere are %d opdets in this event" % opdets.size()
            
    for i_opdet in range(0,opdets.size()):
        opdet = opdets.at(i_opdet)
        #if verbose:
        #    if i_opdet%1000==0: print "\t\tProcessing opdet %d" % i_opdet        
                
        if opdet.TimeStamp()>hirange_us: continue
        if (opdet.TimeStamp()+0.002*opdet.size())<lowrange_us: continue            
                
        this_opdet_df.loc[len(this_opdet_df.index)] = [ run,
                                                        event,
                                                        i_opdet,
                                                        opdet.ChannelNumber(),
                                                        opdet.TimeStamp(),
                                                        np.ndarray(buffer=opdet.data(), 
                                                                   dtype=np.int16, 
                                                                   shape=(opdet.size())
                                                                  )
                                                        ]
    
    opdet_df = opdet_df.append(this_opdet_df)
    if verbose: print "\tdf size is now %d" % len(opdet_df)
    event_counter+=1

    if event_counter%events_per_key==0:
        if verbose: print "Writing to file..."
        #opdet_df.to_hdf(outfile_hdf,key="opdet_df_"+key_number,mode='a',complevel=1)
        opdet_df.to_parquet("/icarus/data/users/wketchum/corsika_detsim_opdet_df_trim"+str(key_number)+".parquet")
        key_number+=1
        opdet_df = pd.DataFrame(columns=df_columns)
    
    ev.next()

if len(opdet_df)!=0:
    if verbose: print "Writing to file..."
    #opdet_df.to_hdf(outfile_hdf,key="opdet_df"+key_number,mode='a',complevel=1)
    opdet_df.to_parquet("/icarus/data/users/wketchum/corsika_detsim_opdet_df_trim"+str(key_number)+".parquet")
    key_number+=1
    opdet_df = pd.DataFrame(columns=df_columns)
```

```python
opdet_df.to_parquet("/icarus/data/users/wketchum/corsika_detsim_opdet_df_trim.parquet")
```
