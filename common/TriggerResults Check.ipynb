```python
filename = b"~/app/dev_areas/mcc9_devs/mcc_9_1_FilterTesting/PhysicsRun-2016_7_30_12_38_37-0007006-00507_20160804T195612_ext_unbiased_20160805T020845_merged_gen_20190510T201952_g4_detsim_mix_r1a_r1b_postdlmctrut_f26f6faa-f2f6-4da4-986c-aa8a507d5189_20190923T143048_ps2fil.root"
filenames = ROOT.vector(ROOT.string)(1,filename)
```

```python
print "Creating event object ..."
ev = gallery.Event(filenames)
```

```python
triggerResults_tag = art.InputTag("TriggerResults","","OverlayFiltersPostStage2");
get_triggerResults = ev.getValidHandle(art.TriggerResults)
```

```python
runs = []
events = []
trigger_paths = ["pi0","NuCC","ncpi0"]
triggers = [ [] for i in range(len(trigger_paths))]

col_names = ["run","event","pi0","NuCC","ncpi0"]
```

```python
ev.toBegin()

print "Entering event loop..."
while (not ev.atEnd()):
    
    run = ev.eventAuxiliary().run()
    event = ev.eventAuxiliary().event()
    print "Run %d, Event %d" % (run,event)

    runs.append(run)
    events.append(event)
    
    results = get_triggerResults(triggerResults_tag)
    for i in range(results.size()):
        triggers[i].append(results.accept(i))
    
    ev.next()
    
print "Done."
```
